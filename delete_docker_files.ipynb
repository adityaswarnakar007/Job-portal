```python
# Remove Docker-related files and verify remaining files
import os
from pathlib import Path

repo_root = Path(r"d:\job")
files = [
    "Dockerfile",
    "docker-compose.yml",
    "docker-entrypoint.sh",
    ".dockerignore",
    "docker_compose_log.txt",
    "docker_output.txt",
]

results = []
for file_name in files:
    path = repo_root / file_name
    try:
        if path.exists():
            path.unlink()
            results.append((file_name, "deleted"))
        else:
            results.append((file_name, "missing"))
    except Exception as exc:
        results.append((file_name, f"error: {exc}"))

remaining = [p.name for p in repo_root.iterdir() if p.name in files]
with open(repo_root / 'docker_delete_results.txt', 'w', encoding='utf-8') as f:
    for result in results:
        f.write(str(result) + '\n')
    f.write('remaining:' + str(remaining) + '\n')
print('done')
```
